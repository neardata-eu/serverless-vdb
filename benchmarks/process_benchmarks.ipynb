```python
import matplotlib.pyplot as plt
import json
```

```python
with open('global_index_gen.json') as f:
    data = json.load(f)
```

```python
# Data for the segments and their corresponding colors
segments = [data['finish_download_vectors'] - data['global_index_start'],
            data['finish_generate_centroids'] - data['finish_download_vectors'], 
            data['finish_generate_csvs'] - data['finish_generate_centroids'],
            data['finish_upload_csvs'] - data['finish_generate_csvs']]  # lengths of segments
colors = ['red', 'green', 'blue', 'yellow']  # colors for segments

for segment in segments:
    print(segment)

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(12, 1))

# Plot a single horizontal bar divided into sections
left = 0
for value, color in zip(segments, colors):
    ax.barh(0, value, color=color, left=left)
    left += value

# Add labels and title
ax.set_yticks([])  # Remove y ticks
ax.set_xlabel('Time (s)')
ax.set_title('Global Index Generation Time Breakdown')
ax.legend(['Download Vectors', 'Generate Centroids', 'Generate CSVs', 'Upload CSVs'], bbox_to_anchor=(1.05, 1), loc='upper left')
# Display the plot
plt.show()
```
